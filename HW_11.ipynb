```python
from sympy import *
from sympy.plotting import plot
x=Symbol('x')
f=0.6*x**3+5.5*x**2+10*x-5
x1,x2,x3 = solve(f,x)
print(x1)
print(x2)
print(x3)
plot(f)
```

```python
f2 = diff(f,x)
print(f2)
x_1,x_2 = (solve(f2,x))
print(x_1)
print(x_2)
print(f"Области возрастания [- беск ; {x_1}] и [{x_2} ; + беск]")
print(f"Область убывания [{x_1};{x_2}])")
```
